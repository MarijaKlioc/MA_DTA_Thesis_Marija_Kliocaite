```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

confusion_matrices = {
    "anticipation": np.array([[2787,   47], [ 382,   43]]),
    "optimism": np.array([[1850,  266], [ 408,  735]]),
    "trust": np.array([[3090,   16], [ 145,    8]]),
    "joy": np.array([[1668,  149], [ 340, 1102]]),
    "love": np.array([[2604,  139], [ 257,  259]]),
    "anger": np.array([[1937,  221], [ 273,  828]]),
    "disgust": np.array([[1885,  275], [ 273,  826]]),
    "pessimism": np.array([[2833,   51], [ 340,   35]]),
    "sadness": np.array([[2118,  181], [ 393,  567]]),
    "fear": np.array([[2660,  114], [ 183,  302]]),
    "surprise": np.array([[3076,   13], [ 154,   16]])
}

# Set up the figure with a main title and subplots for each confusion matrix
fig, axes = plt.subplots(2, 6, figsize=(16, 5))

# Flatten axes for easier iteration
axes = axes.flatten()

# Create heatmaps
for ax, (emotion, cm) in zip(axes, confusion_matrices.items()):
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", ax=ax, cbar=False, annot_kws={"size": 12})
    ax.set_title(emotion, fontsize=14)
    ax.set_xlabel('Predicted', fontsize=12)
    ax.set_ylabel('True', fontsize=12)
    ax.tick_params(axis='both', which='major', labelsize=9)

# Remove any unused subplots
for ax in axes[len(confusion_matrices):]:
    fig.delaxes(ax)

# Adjust layout to prevent overlap
plt.tight_layout(rect=[0.03, 0, 1, 0.95])

# Save the figure with proper margins
plt.savefig('confusion_matrices_gpt_increased_sl_roberta.png', dpi=300, bbox_inches='tight', pad_inches=0.5)

# Show the plot
plt.show()
```
